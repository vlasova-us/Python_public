```python
import pandas as pd
from datetime import datetime
```

```python
# Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

df_movies = pd.read_csv('/content/u.item.csv', sep = '|', encoding='latin1', header=None,
                        names = ['movie_id', 'movie_title', 'release_date', 'video_release_date',
                                 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens',
                                 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
                                 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller ', 'War',
                                 'Western'])
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], format='%d-%b-%Y')

df_movies
```

```python
df_ratings = pd.read_csv('/content/u.data.csv', sep = '\t', encoding='latin1', header=None,
                         names = ['user_id', 'item_id', 'rating', 'timestamp'])
df_ratings
```

```python
# Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

df_bigest_rating_number = df_ratings.groupby(['user_id']).count()[['rating']].reset_index().sort_values(by='rating', ascending=False).head(1)
df_bigest_rating_number
```

```python
# Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь

df_user_405 = df_ratings[df_ratings['user_id'] == 405]
df_user_405
```

**Добавьте к датафрейму из задания 3 столбцы:**
# По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет
# столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей

```python
# подсчет общего количества оценок по каждому фильму

df_total_count_rattings = df_ratings.groupby(['item_id']).count()[['rating']].reset_index()
df_total_count_rattings.rename(columns = {'rating': 'total_count_rattings'}, inplace = True )

df_total_count_rattings
```

```python
# подсчет общей суммы оценок по каждому фильму

df_total_sum_rattings = df_ratings.groupby(['item_id']).sum()[['rating']].reset_index()
df_total_sum_rattings.rename(columns = {'rating': 'total_sum_rattings'}, inplace = True )
df_total_sum_rattings
```

```python
# Объединение df

df_user_405_genre = pd.merge(df_user_405, df_movies, how = 'left', left_on = 'item_id', right_on = 'movie_id')
df_user_405_genre_count_rattings = pd.merge(df_user_405_genre, df_total_count_rattings, how = 'left', left_on = 'item_id', right_on = 'item_id')
df_for_analysis = pd.merge(df_user_405_genre_count_rattings, df_total_sum_rattings, how = 'left', left_on = 'item_id', right_on = 'item_id')
df_for_analysis
```

```python
# Формируем X_train, X_test, y_train, y_test

from sklearn.model_selection import train_test_split
df_for_analysis.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_for_analysis[['movie_id',
       'unknown', 'Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller ', 'War', 'Western',
       'total_count_rattings', 'total_sum_rattings']], df_for_analysis['rating'], test_size=0.2)
```

```python
# Приводим данные к одному масштабу при помощи StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Обучаем модель
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test)
```

```python
# Оцениваем качество, при помощи метода mean_squared_error для тестовой выборки
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, lr.predict(X_test))
```

```python
# разница факта и прогноза
result = pd.DataFrame([y_test, lr.predict(X_test)]).T
result
```

```python
# расчет точности
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lr.predict(X_test))
```

```python
lr.score(X_test, y_test)
```
