```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 11,7
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data/googleplaystore.csv')
df
```

```python
# удаляю дубликаты игр
google_play_df = df.drop_duplicates(subset=['App'])
google_play_df
```

```python
# обнаружена аномальная строка, в которой отсутствует значение Category и все данные по колонкам смещены. Восстановить значение Category не возможно.

google_play_df.drop(labels = [10472],axis = 0, inplace = True)
```

```python
# Преобразование типов данных
google_play_df['Reviews'] = google_play_df['Reviews'].astype('int')
google_play_df['Price']= google_play_df['Price'].apply (lambda x: x.replace(',', '').replace('$', '')).astype('float')
google_play_df['Last Updated'] = pd.to_datetime(google_play_df['Last Updated'], format = '%B %d, %Y', dayfirst=True)
```

```python
google_play_df.info()
```

```python
# Большинство приложений в GooglePlay имеет рейтинг выше 4
google_play_df['Rating'].plot(kind='hist', 
                   title='Распределение оценок рейтинга приложений', bins=20)
```

```python
# По количеству приложений в каждой можно оценить цифровизацию разных сторон жизни. Из данного набора данных можно сделать вывод, что большинство приложений в GooglePlay  - это инструменты, разлекательные и образовательные программы.
top10_app_in_genres = google_play_df[['Genres', 'App']].groupby('Genres').count().sort_values('App', ascending=False).head(10)
top10_app_in_genres.plot(kind='bar', title='Топ 10 жанров по количество приложений',
                             xlabel='Жанры',
                             ylabel='Количество приложений')
for i, val in enumerate(top10_app_in_genres.values):
    plt.text(i, val, int(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'size':14})
```

```python
# Корреляция между Просмотрами, Рейтингом и ценой. У приложений с рейтингом от 4 до 5 есть взаимосвязь с количеством просмотров в GooglePlay
sns.pairplot(google_play_df[['Rating', 'Price', 'Reviews']])
```

```python
# Большинство приложений бесплатны
pie_data = google_play_df['Type'].value_counts()
pie_data.plot(kind='pie', title='Количество платных и бесплатных приложений')
plt.legend()
```
