**Задание 1**

Для датафрейма log из материалов занятия создайте столбец source_type по правилам:

если источник traffic_source равен Yandex или Google, то в source_type ставится organic;

для источников paid и email из России ставим ad;

для источников paid и email не из России ставим other;

все остальные варианты берём из traffic_source без изменений.


```python
import pandas as pd
table = pd.read_csv ('/content/drive/MyDrive/Data/visit_log.csv', sep = ';')

def traffic (row):
 ''' Правила заполнения столбца source_type:
- если источник traffic_source равен Yandex или Google, то в source_type ставится organic;
- для источников paid и email из России ставим ad;
- для источников paid и email не из России ставим other;
- все остальные варианты берём из traffic_source без изменений. '''
 
 if ('yandex' in row ['traffic_source']) or ('google' in row ['traffic_source']):
  return 'organic'

 if ('paid' in row ['traffic_source'] and 'Russia' in row ['region']) or ('email' in row ['traffic_source'] and 'Russia' in row ['region']):
  return 'ad'

 if ('paid' in row ['traffic_source'] and 'Russia' not in row ['region']) or ('email' in row ['traffic_source'] and 'Russia' not in row ['region']):
  return 'other'

 return row ['traffic_source']
  
table ['source_type'] = table.apply (traffic, axis = 1)
table.head()
```

**Задание 3**

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.




```python
average_time_living = pd.read_csv ('/content/drive/MyDrive/Data/ml-latest-small/ratings.csv', sep = ',')

def max_min(column):
 '''В функцию подается столбец timestamp, сгруппированный по UserId, для вычисления разницы между максимальным и минимальным значением по столбцу timestsmp для данного UserId'''
 return column.max() - column.min() 

average_time_living.groupby('userId').agg(['count', max_min])['timestamp'].query('count >100').mean().head()
```

**Задание 4**
Дана статистика услуг перевозок клиентов компании по типам.
Нужно сформировать две таблицы:

- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;
- аналогичную таблицу по типам выручки с указанием адреса клиента.

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# таблица с тремя типами выручки для каждого client_id без указания адреса клиента

table_concat = pd.concat([rzd, auto, air], axis=0).groupby('client_id').sum().head(10)
table_concat
```

```python
# аналогичная таблица по типам выручки с указанием адреса клиента.
revenue_adress_table = table_concat.merge (client_base, on = 'client_id', how = 'left')
revenue_adress_table
```
