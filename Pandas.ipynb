Дан датафрейм Pandas Sales, в котором хранятся продажи и возвраты:

|    Shop    |   Employee   |     Type     |    Sum    |

Описание полей:

* Shop - название магазина
* Employee - имя сотрудника
* Type - тип транзакции (продажа - 1, возврат - 2)
* Sum - сумма транзакции

Руководитель просит подбить итоги в разрезе каждого сотрудника в каждом отдельном магазине:

* Сколько возвратов и сколько продаж было?
* Какая сумма продаж и возвратов?
* Какая средняя сумма продажи и возврата?
* Результат необходимо оформить в виде датафрейма следующего вида:

|    Shop    |   Employee   |     Type     |    Amount    |    Sum    |      Avg      |

Примечание: После всех преобразований все поля (Shop, Employee, Type, Amount, Sum, Avg) должны быть столбцами итогового датафрейма df. Другими словами, например, конструкция df['Shop'] не должна выдавать ошибку.



```python
import pandas as pd

l = [
    {'Shop': "Yakimanka", 'Employee': "Anna", 'Type': 1, 'Sum': 528},
    {'Shop': "Yakimanka", 'Employee': "Fedor", 'Type': 2, 'Sum': 632},
    {'Shop': "Butovo", 'Employee': "Peter", 'Type': 1, 'Sum': 115},
    {'Shop': "Yakimanka", 'Employee': "Anna", 'Type': 1, 'Sum': 1024},
    {'Shop': "Butovo", 'Employee': "Peter", 'Type': 1, 'Sum': 1754},
    {'Shop': "Butovo", 'Employee': "Peter", 'Type': 1, 'Sum': 111},
    {'Shop': "Kamenka", 'Employee': "Max", 'Type': 1, 'Sum': 499},
    {'Shop': "Kamenka", 'Employee': "Max", 'Type': 2, 'Sum': 23},
    {'Shop': "Butovo", 'Employee': "Jugen", 'Type': 1, 'Sum': 214},
    {'Shop': "Kamenka", 'Employee': "Max", 'Type': 1, 'Sum': 964},
    {'Shop': "Yakimanka", 'Employee': "Fedor", 'Type': 2, 'Sum': 100},
    {'Shop': "Butovo", 'Employee': "Jugen", 'Type': 1, 'Sum': 10000}
]

df = pd.DataFrame(l)
df
```

```python
df = df.groupby(['Shop', 'Employee', 'Type'], as_index=False).agg({'Sum': ['count', 'sum', 'mean']})
df.columns = ['_'.join(col).rstrip('_') for col in df.columns.values]
df.columns = ['Shop', 'Employee', 'Type', 'Amount', 'Sum', 'Avg']
df



```

Дан датафрейм Pandas, состоящий из 1 столбца - sales. В нем содержатся продажи за некоторый период времени: отдельная строка - отдельный период времени.

Необходимо в исходный датафрейм добавить 2 столбца:

* prev_sales - продажи за предыдущий период
* inc - рост продаж по сравнению с предыдущим периодом

Данное преобразование необходимо совершать с помощью функции evalInc. На вход подается исходный датафрейм, функция должна возвращать преобразованный датафрейм.

Примечание: Значения в столбце inc должны быть выражены в % (т.е. не 0.01, а 1) и округлены до 2 знаков после запятой.

Примечание: Если для данной записи нет предыдущего значения, то prev_sales должно быть NaN.

```python
import pandas as pd
import numpy as np

d = {'sales': [100, 150, 150, 120, 135, 180, 230, 500, 501, 499, 500, 500]}
df = pd.DataFrame(d)
```

```python
def evalInc(df):
  df['prev_sales'] = df['sales'].shift(1,fill_value=0)
  df['inc'] = (df['sales']*100/df['prev_sales'] - 100).round(2)
  df['inc'].replace([np.inf , -np.inf ], np.nan , inplace= True )

  return df

evalInc (df)
```
